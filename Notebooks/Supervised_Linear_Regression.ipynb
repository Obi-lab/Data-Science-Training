```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example dataset: Square footage (X) and house prices (y)
X = np.array([[500], [800], [1000], [1200], [1500], [1800]])  # Square footage
y = np.array([100000, 150000, 180000, 200000, 240000, 300000])  # Prices in USD

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)  # Train the model

# Predict prices for new square footage
new_sqft = np.array([[1100], [1400], [1600]])
predictions = model.predict(new_sqft)

# Visualize the results
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, model.predict(X), color='red', label='Linear Regression Line')
plt.scatter(new_sqft, predictions, color='green', label='Predictions')
plt.xlabel('Square Footage')
plt.ylabel('House Price (USD)')
plt.legend()
plt.show()
```
